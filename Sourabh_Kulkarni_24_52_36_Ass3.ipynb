```python
import numpy as np
import matplotlib.pyplot as plt

# Linear regression using Normal Equation
def normal_equation(X, y):
    # Add a column of ones to X for the intercept term (bias)
    X_bias = np.c_[np.ones((X.shape[0], 1)), X]

    # Apply the normal equation: theta = (X^T * X)^(-1) * X^T * y
    theta = np.linalg.inv(X_bias.T @ X_bias) @ X_bias.T @ y
    return theta

# Generate input data (features) and target values
np.random.seed(42)  # For reproducibility
X = np.linspace(0, 10, 50).reshape(-1, 1)
y = 3 * X + np.random.randn(50, 1) * 2  # True slope = 3, noise added

# Get the MLE theta using the normal equation
theta = normal_equation(X, y)

# Predict the values using the learned parameters
X_bias = np.c_[np.ones((X.shape[0], 1)), X]
y_pred = X_bias @ theta

# Plot the original data and the predicted line
plt.scatter(X, y, color='blue', label='Original Data')
plt.plot(X, y_pred, color='red', label='Linear Regression Prediction')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Linear Regression using Normal Equation')
plt.show()
```
